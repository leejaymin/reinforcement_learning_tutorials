```python
# 함수근사 방법
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
import matplotlib.pyplot as plt
```

```python
plt.scatter(diabetes.data[:,2], diabetes.target)
plt.ylabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = diabetes.data[:,2]
y = diabetes.target
```

```python
w = 1.0
b = 1.0
# y_hat = wx + b (linear function)
y_hat = x[0]*w + b
print(y_hat, y[0])
```

```python
err = y[0] - y_hat
w_new = w + x[0]*err
b_new = b + 1*err
```

```python
y_hat = x[0] * w_new + b_new
print(y_hat, y[0])
```

```python
y_hat = x[1] * w_new + b_new
print(y_hat, y[1])
```

```python
w = b = 1.0
cnt = 0
for i in range(1, 2):
    for x_i, y_i in zip(x,y):
        cnt += 1
        y_hat = x_i * w + b
        err = y_i - y_hat
        w_rate = x_i
        w = w + w_rate * err
        b = b + err
        if cnt % 100 == 0:
            print(w, b)
            plt.scatter(x,y)
            pt1 = (-0.1, -0.1*w+b)
            pt2 = (0.15, 0.15*w+b)
            plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
            plt.ylabel('x')
            plt.ylabel('y')
            plt.show()

print(w,b)
```

```python
plt.scatter(x,y)
pt1 = (-0.1, -0.1*w+b)
pt2 = (0.15, 0.15*w+b)
plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]])
plt.ylabel('x')
plt.ylabel('y')
plt.show()
```

```python
lr = 0.1
bias =1
weight = [0.4, 0.1, -0.3]
def Perceptron(input1, input2, output=1):
    outputP = input1*weight[0] + input2*weight[1] + bias * weight[2]

    if outputP > 0:
        outputP = 1
    else:
        outputP = 0

    if output == -1:
        return outputP

    error = output - outputP
    weight[0] += error * input1 * lr
    weight[1] += error * input2 * lr
    weight[2] += error * bias * lr
for _ in range(10):
    Perceptron(1,1,1)
    Perceptron(1,0,0)
    Perceptron(0,1,0)
    Perceptron(0,0,0)
    print(weight)

print(Perceptron(1,1), \
Perceptron(1,0), \
Perceptron(0,1), \
Perceptron(0,0))
```
